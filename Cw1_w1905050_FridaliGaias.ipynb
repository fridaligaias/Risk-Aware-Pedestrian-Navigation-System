The transportation sector is currently undergoing a transformation driven by Artificial Intelligence, like in many, if not arguably all, fields.

Today, AI is applied across the entire transport system, from controlling a single autonomous vehicle to the wider managing of city wide traffic flows and global logistics networks. My review will cover how some AI methodologies are applied across three sub-fields in Transportation: Autonomous Vehicles, Intelligent Traffic Systems (ITS), and Route Optimisation.

**Autonomous Vehicles**

To navigate safely, an autonomous vehicle must first perceive and understand its surrounding environment (Badue et al., 2021). Convolutional Neural Networks (CNNs) are the dominant approach for semantic segmentation/object detection in autonomous driving systems, enabling the classification of pedestrians, lane markings, traffic signs and so on, from raw pixel data. Some examples of development in these methods include the works of Long, Shelhamer and Darrell, 2015 and Chen et al., 2018.
By learning hierarchical features from simple edges to complex object representations, CNNs allow vehicles to interpret road scenes in real time (LeCun, Bengio and Hinton, 2015).

Once perception is achieved, autonomous vehicles must make safe and concrete decisions. While reinforcement learning is commonly applied, "game-theoretic" and minimax based planning approaches are particularly useful in interactive scenarios involving human drivers. In such cases, lane changing and intersection merging can be modelled as strategic games, where the autonomous agent seeks to maximise safety under worst-case assumptions about other agents' behaviour (Fisac et al., 2019).

**Route Optimisation**

Optimising the movement of goods and people addresses problems such as the Shortest Path Problem and the Vehicle Routing Problem (VRP), which have been extensively studied over the years. Modern navigation systems, including GPS-based routing applications, rely on AI search techniques to compute efficient routes through graph representations of road networks (Russell and Norvig, 2021).  While classical search algorithms typically assume static costs such as distance, real world transportation systems increasingly rely on predictive machine learning to model more dynamic conditions. Regression models are now commonly used to predict travel times based on historical traffic data and contextual variables such as time of day, weather conditions, and day of the week (Vlahogianni, Karlaftis and Golias, 2014; Wang et al., 2016). This mixed approach, where machine learning models estimate costs and search algorithms compute optimal routes, enable more accurate and reliable arrival time predictions in intelligent transportation systems.

**Intelligent Transporation Systems**

Intelligent Transporation Systems (ITS) aim to optimise city wide traffic flow and reduce congestion through the use of data driven decision making and adaptive control strategies (Vlahogianni, Karlaftis and Golias, 2014). Clustering algorithms are used to identify traffic patterns and congestion hotspots by grouping road segments with similar traffic characteristics. This allows traffic management centres to categorise areas as free flowing, congested, or incident prone, enabling more targeted and dynamic traffic control interventions (Sun, S. et al., 2012; Lu, et al., 2012).
To model the non-linear and temporal nature of traffic flow, ITS increasingly rely on neural networks. Recurrent Neural Networks (RNNs) are  good for analysing time series data collected from detectors and roadside sensors (Ma et al., 2015). These models enable accurate prediction of future congestion levels, allowing traffic signals and control systems to adapt proactively rather than responding reactively to congestion once it has already formed.

## Part B: Compare and Evaluate AI Techniques (12 Marks)

**Project Goal**: To develop a pedestrian navigation system ("SafeWalk") that calculates the optimal walking route by balancing travel distance with personal safety risk derived from local historical crime data.

To achieve this goal, I'll evaluate three different AI techniques taught in this module: A* Search, Unsupervised Learning (K- Means Clustering), and Artificial Neural Networks (ANNs).

1- A* Search Algorithm

Search algorithms are great for problem solving where the goal is to reach a specific state. A* Search is a "Best-First" strategy that combines the path cost so far $(g(n))$ with a heuristic estimate to the goal $(h(n))$ (Basukoski, 2025).

An example I found during research is described by Calkins (2014), who details how the A* algorithm is used in game development to navigate units across variable terrain costs. Similar to how Calkins assigns higher "movement costs" to difficult terrain (like swamps or mountains) to make agents avoid them, for "SafeWalk," I could adapt the cost function $f(n) = g(n) + h(n)$ so that $g(n)$ represents a combined weight of distance and crime risk, allowing the agent to "search" for the best path.



Strengths: A* is complete (it will always find a solution if one exists) and optimal (it finds the best solution), provided the heuristic is admissible. A* is particularly effective because it directs the search toward the goal rather than exploring blindly, making it computationally efficient for routing compared to other search methods.

Weaknesses: A major drawback is memory consumption, as the algorithm must keep all generated nodes in memory. Furthermore, it assumes a static environment; it does not inherently learn or adapt to new crime trends without manual weight updates.



Input: A map state space and a defined start and goal (destination) state.

Output: A sequence of actions (path) from start to goal.

---

2- Unsupervised Learning: K-Means Clustering

Machine Learning involves systems that learn from data. Unsupervised Learning is a method where "the system tries to learn without a teacher" and the data has no associated target attribute.

K-Means is good for data analysis and finding "hidden" patterns in data, it can be applied for feature engineering  (Bonmati, Lecture 7).

Wahyono (2024) utilized K-Means clustering to identify traffic accident hotspots in urban environments. By grouping historical accident coordinates, the system successfully categorized city districts into distinct risk levels (Blue = Subdistricts with moderate accident rates, Green = Subdistricts with low accident rates, Orange = Subdistricts with high accident rates) to inform local authorities.

A similar approach could be used to cluster crime incidents instead of traffic accidents, to analyse the spatial distribution of crime. By grouping crime coordinates into clusters, high density "danger zones" could be found to warn and guide users.



Strengths: As aforementioned, clustering is excellent for finding hidden structures in data. It could automatically categorise crime maps into "risk regions" (e.g., Safe vs. Dangerous).

Weaknesses: Result depend heavily on the chosen inputs (bias/variance). Crucially for this project, clustering is descriptive, not prescriptive; it can identify a bad area, but it cannot calculate a valid path around it. It would therefore need a separate search algorithm to actually navigate, so doesn't seem that efficient.



Input: Unlabeled training data (crime coordinates).

Output: Grouped subsets (clusters) of data.


---

3- Artificial Neural Networks

ANNs are systems inspired by biological neural networks, consisting of interconnected layers of nodes. They have been successfully applied to a wide range of engineering problems and have increasingly been explored within the field of transportation. The big advantage that ANNs have is that they learn directly from example data rather than requiring each case to be explicitly modelled mathematically (Çetiner et al., 2010). This makes them useful in the context of complex and safety related transport problems where relationships between variables are non linear and difficult to define in advance.

Çetiner et al. (2010) applied ANNs to predict traffic flow volume at specific junctions. Their model successfully learned patterns from historical data to forecast future congestion states. A similar architecture could be applied to my project to predict "crime risk" at a specific street/area based on time of day, rather than traffic volume.



Strengths: ANNs are powerful for modeling non linear functions where the relationship between inputs and outputs is complex (Bonmati, Lecture 8). They would very likely offer good predictive accuracy for future crimes rather than simple historical averages.

Weaknesses: The "black box" problem. It makes it difficult to explain why a route was chosen, an issue for user trust in real life applications. Additionally, training requires a large volume of data and significant computational time to adjust weights via backpropagation (Bonmati, Lecture 8), making them less suitable for a prototype implementation.



Input: A large dataset of inputs (features) and target outputs (labels) for supervised training.

Output: A predicted value (risk score) or classification.

---

For the implementation of my "SafeWalk" prototype, I have selected A* Search as the core technique.

Unlike ANNs, which obscure decision logic, A* allows the cost function to be explicitly defined and inspected. This ensures that routing decisions are transparent, verifiable, and mathematically grounded. Compared to clustering, A* directly produces actionable routes rather than descriptive insight. Given the constraints of a prototype and the importance of ethics in AI and especially, I believe, in safety focused navigation, A* heuristic search offers the most appropriate balance between performance, interpretability, and implementation complexity.

Crime type: A categorical String description (e.g. "Violence and sexual offences", "Theft", etc.).

---

2- Data Pre-processing

Raw crime data cannot be directly used by the A* algorithm, it requires three stages of pre-processing (implemented in `Block 3`, `Block 4`, `Block 5`):

* "Bounding Box" filter: To optimise memory usage and processing times, the wider crime dataset is filtered to include only incidents falling within the min/max latitude and longitude of the given map location.

* Severity Mapping: The A* algorithm requires numerical weights, but the input data provides categorical labels. A `severity_map` dictionary is applied to transform crime types into weighted integers. High harm crimes (e.g., "Violence and sexual offences") are weighted more heavily than other crimes (e.g., "Other theft").

* Node Matching: Crimes are spatially joined to the graph (`ox.nearest_nodes`, (Engineering Code Hub)). The severity scores are summed for each node, resulting in a single `risk_score` attribute for every intersection on the map.

```python
#block 2
# User config, setting Alpha value

import os
import osmnx as ox
import networkx as nx
import pandas as pd
import numpy as np
import folium
from geopy.distance import geodesic
import matplotlib.pyplot as plt

#defining initial variables: area + crime data
PLACE_NAME = "Marylebone, London, United Kingdom"
#crime data
CRIME_DATA_PATH = os.path.join(os.getcwd(), "metMerged_crime_data.csv")

#setting alpha: lower alpha--> A* shortest path to goal, higher alpha--> danger heavily weighted in cost function, dangerous areas avoided more!
while True:
    try:
        ALGORITHM_ALPHA = int(input("Please enter a number between 0 and 100: "))
        if 0 <= ALGORITHM_ALPHA <= 100:
            print(f"You entered: {ALGORITHM_ALPHA}")
            break
        else:
            print("Error! The number must be between 0 and 100.")
    except ValueError:
        print("Invalid input. Please enter a whole number.")

print(f"Alpha value set to: {ALGORITHM_ALPHA} for this test.")
```

```python
# block 3 - data loading
print(f"Downloading street network for wider {PLACE_NAME} area...")

#define the center point of (PLACE_NAME), coordinates for Marylborne:
center_lat, center_lon = 51.51908373970027, -0.1505046148528704

#get graph within a distance (dist) in meters
G = ox.graph_from_point((center_lat, center_lon), dist=2000, network_type='all_public')

G_proj = ox.project_graph(G)

#setting the bounds
nodes = ox.graph_to_gdfs(G, nodes=True, edges=False)
min_lon, min_lat, max_lon, max_lat = nodes.total_bounds

print(f"Graph loaded! Nodes: {len(G.nodes)}, Edges: {len(G.edges)}")
```

```python
#block 4 - attaching severity to crimes

def load_crime_data(filepath, bbox):
    min_x, min_y, max_x, max_y = bbox

    df = pd.read_csv(filepath)
    print("Crime data loaded.")
    #keep only crimes within set boundaries:
    df = df[
        (df['Latitude'] >= min_y) & (df['Latitude'] <= max_y) &
        (df['Longitude'] >= min_x) & (df['Longitude'] <= max_x)
    ]

    #assigning weights to crimes based on severity
    #high score = high danger
    severity_map = {
        'Violence and sexual offences': 10,
        'Possession of weapons': 9,
        'Criminal damage and arson': 9,
        'Theft from the person': 8,
        'Robbery': 8,
        'Drugs': 7,
        'Anti-social behaviour': 7,
        'Public order': 5,
        'Burglary': 5,
        'Criminal damage and arson': 5,
        'Shoplifting': 4,
        'Bicycle theft': 3,
        'Other theft': 2,
        'Vehicle crime': 2,
        'Other crime': 2,
    }

    # mapping severity, default to 1 if not in list (if for any reason, missing crime type)
    df['severity'] = df['Crime type'].map(severity_map).fillna(1)

    return df

crime_df = load_crime_data(CRIME_DATA_PATH, (min_lon, min_lat, max_lon, max_lat))
print(f"Crimes within area: {len(crime_df)}")
display(crime_df.head())
```

```python
#block 5 - risk scores to nodes

#setting risk score on all nodes to 0 initially
nx.set_node_attributes(G, 0, 'risk_score')

print("Mapping crimes to nearest street nodes ...")

nearest_nodes = ox.nearest_nodes(G, crime_df['Longitude'], crime_df['Latitude'])

crime_df['nearest_node'] = nearest_nodes

# index = Node ID and value = Total Risk
node_risks = crime_df.groupby('nearest_node')['severity'].sum()

#Series to dictionary and map
risk_dict = node_risks.to_dict()
nx.set_node_attributes(G, risk_dict, 'risk_score')

print("Risk mapping finished!")
```

```python
#block 6 - [adapted from Module Materials - Week 4 Search
# ref: University of Westminster (2024) Tut4Search.ipynb. [Jupyter Notebook] Module Materials – Week 4. University of Westminster
# (accessed December 2025).]
import math
import heapq

# abstract Problem class
class Problem(object):
    def __init__(self, initial=None, goal=None, **kwds):
        self.__dict__.update(initial=initial, goal=goal, **kwds)

    def actions(self, state): raise NotImplementedError
    def result(self, state, action): raise NotImplementedError
    def is_goal(self, state): return state == self.goal
    def action_cost(self, s, a, s1): return 1
    def h(self, node): return 0

#node slass
class Node:
    def __init__(self, state, parent=None, action=None, path_cost=0):
        self.state = state
        self.parent = parent
        self.action = action
        self.path_cost = path_cost
        self.depth = 0
        if parent:
            self.depth = parent.depth + 1

    def __repr__(self): return '<Node {}>'.format(self.state)
    def __lt__(self, other): return self.path_cost < other.path_cost

    def expand(self, problem):
        return [self.child_node(problem, action)
                for action in problem.actions(self.state)]

    def child_node(self, problem, action):
        next_state = problem.result(self.state, action)

        new_cost = self.path_cost + problem.action_cost(self.state, action, next_state)
        next_node = Node(next_state, self, action, new_cost)
        return next_node

    def path(self):
        node, path_back = self, []
        while node:
            path_back.append(node)
            node = node.parent
        return list(reversed(path_back))

#PriorityQueue
class PriorityQueue:
    def __init__(self, items=(), key=lambda x: x):
        self.key = key
        self.items = []
        for item in items: self.add(item)

    def add(self, item):
        heapq.heappush(self.items, (self.key(item), item))

    def pop(self): return heapq.heappop(self.items)[1]
    def __len__(self): return len(self.items)

#Search algo
def best_first_search(problem, f):
    node = Node(problem.initial)
    frontier = PriorityQueue([node], key=f)
    reached = {problem.initial: node}

    while frontier:
        node = frontier.pop()
        if problem.is_goal(node.state):
            return node

        for action in problem.actions(node.state):
            s = problem.result(node.state, action)

            new_cost = node.path_cost + problem.action_cost(node.state, action, s)
            child = Node(s, node, action, new_cost)

            if s not in reached or child.path_cost < reached[s].path_cost:
                reached[s] = child
                frontier.add(child)
    return None

def astar_search(problem, h=None):
    #f(n) = g(n) + h(n)
    h = h or problem.h
    return best_first_search(problem, f=lambda n: n.path_cost + h(n))

print("A* AI Engine loaded!")
```

```python
#block 7
#subclassing Problem to SafeWalk [adapted from Module Materials - Week 4 Search ref: , from prob-->A, inherits Problem class.*]
# ref: University of Westminster (2024) Tut4Search.ipynb. [Jupyter Notebook] Module Materials – Week 4. University of Westminster
# (accessed December 2025).]

class SafeWalkProblem(Problem):
    def __init__(self, initial, goal, graph, alpha=50):
        super().__init__(initial, goal)
        self.graph = graph
        self.alpha = alpha  #risk

    def actions(self, state):
        return list(self.graph.neighbors(state))

    def result(self, state, action):
        return action

    def action_cost(self, s, action, s1):
        #Cost = Distance + (Risk * Alpha)
        # OSMnx graphs store data in edges --> edge between s and s1
        edge_data = self.graph.get_edge_data(s, s1)[0]
        dist = edge_data.get('length', 1)

        risk = self.graph.nodes[s1].get('risk_score', 0)

        return dist + (risk * self.alpha)

    def h(self, node):
        # straight-line heuristic to the goal
        n_node = self.graph.nodes[node.state]
        g_node = self.graph.nodes[self.goal]

        return geodesic(
            (n_node['y'], n_node['x']),
            (g_node['y'], g_node['x'])
        ).meters

print("SafeWalk loaded!")
```

```python
#block 7 - execution based on path

#getting coordinates
#preset path:
# start: University of Westminster - Cavendish Campus
#origin_point = (51.52105483574651, -0.1397713381813016)
# goal (destination) : The National Gallery
#destination_point = (51.50937121732973, -0.13280961717881726)

#user defined coordinates for origin to destination path:
def get_coord(prompt, min_value, max_value):
  while True:
    try:
      value = float(input(prompt))
      if min_value <= value <= max_value:
        return value
      else:
        print(f"Error! Value must be between {min_value} and {max_value}.")
    except ValueError:
      print("Invalid input... Please enter a valid number!")
print("---Enter Origin Coordinates: ---")
origin_lat = get_coord("Enter Latitude: ", -90, 90)
origin_lon = get_coord("Enter Longitude: ", -180, 180)

origin_point = (origin_lat, origin_lon)

print("\n---Enter Destination Coordinates: ---")
dest_lat = get_coord("Enter Latitude: ", -90, 90)
dest_lon = get_coord("Enter Longitude: ", -180, 180)

destination_point = (dest_lat, dest_lon)
print("\n---Your path: ---")
print(f"Origin: {origin_point}")
print(f"Destination: {destination_point}")

# node ids from map
orig_node = ox.distance.nearest_nodes(G, origin_point[1], origin_point[0])
dest_node = ox.distance.nearest_nodes(G, destination_point[1], destination_point[0])

#connect to problem
safe_problem = SafeWalkProblem(initial=orig_node, goal=dest_node, graph=G, alpha=ALGORITHM_ALPHA)

print("\nStarting SafeWalk A* Search...")

#run a*
result_node = astar_search(safe_problem)

if result_node:
    #extracting the path from the nodes
    path_nodes = result_node.path()
    route_safest = [n.state for n in path_nodes] #back to ID list

    print(f"Destination found!")
    print(f"\nSafest Route Nodes: {len(route_safest)}...")

    #run standard (for comparison)
    route_shortest = nx.shortest_path(G, orig_node, dest_node, weight='length')
    print(f"Shortest Route Nodes: {len(route_shortest)}...")


    length_short = nx.path_weight(G, route_shortest, weight='length')
    length_safe = nx.path_weight(G, route_safest, weight='length')

    print(f"\nShortest Route Length: {length_short:.1f} m")
    print(f"Safest Route Length: {length_safe:.1f} m")

    print(f"\nSafest route has a node cost of {len(route_safest) - len(route_shortest)}, and is {int(length_safe - length_short)}m longer.")

else:
    print("No path found.")
```

## Part D: Software Testing and Evaluation. (20 Marks)

### 1. Testing Methodology and Verification

To verify the correctness and usability of my A* implementation, I used a combination of comparative testing and analysis.

#### Method 1: Correctness

Standard navigation relies on minimizing physical distance. To verify my custom A* algorithm, I needed to ensure my heuristic is admissible, so I compared its output against a baseline: the standard NetworkX `shortest_path`, and `astar_path_length` functions.

**Test Case 1 (Alpha = 0)**: When the risk parameter (`ALGORITHM_ALPHA`) is set to 0, the cost function effectively becomes $f(n) = g(distance) + h(distance)$.

**Expected Result**: The "Safest Route" must be identical to the "Shortest Route."

**Actual Result**: In testing, setting Alpha to 0 resulted in the Blue (Safe) and Red (Shortest) lines overlapping perfectly, with a length difference of 0 meters. This confirms the underlying graph traversal logic is mathematically correct and the heuristic is admissible.

References:

* NetworkX 'Shortest Paths' NetworkX 3.6.1 documentation. Available at: https://networkx.org/documentation/stable/reference/algorithms/shortest_paths.html (Accessed: December 2025)

* NetworkX 'shortest_path_length' NetworkX documentation. Available at: https://pelegm-networkx.readthedocs.io/en/latest/reference/generated/networkx.algorithms.shortest_paths.generic.shortest_path_length.html
 (Accessed: December 2025)

```python
#checking heuristic admissibility
print("--- Test Case 1 ---")

# running SafeWalk with Alpha = 0 (no risk penalty)
# --> acts like a standard distance search
test_problem = SafeWalkProblem(orig_node, dest_node, G, alpha=0)
node_result = astar_search(test_problem)
path_a_star_zero = [n.state for n in node_result.path()]

#measuring lengths
len_astar_zero = nx.path_weight(G, path_a_star_zero, weight='length')
len_base = nx.shortest_path_length(G, orig_node, dest_node, weight='length') #default NetworkX's shortest path
lena_base = nx.astar_path_length(G, orig_node, dest_node, weight='length') #NetworkX's A* search path

print(f"Standard Base Path Length: {len_base:.4f} m")
print(f"A* Path Length: {lena_base:.4f} m")
print(f"My own A* (Alpha=0) Path Length: {len_astar_zero:.4f} m")

diff = abs(len_astar_zero - len_base and len_astar_zero - lena_base)
if diff == 0:
    print("\nTEST PASSED!")
else:
    print(f"\nTEST FAILED! Difference of {diff} meters...")
```

#### Method 2: Effectiveness

I tested the algorithm's responsiveness by incrementally increasing the Alpha value (0, 30, 50, 100).

**Expected Result**: As Alpha increases, the path length should increase or stay the same (never decrease), while the cumulative "Risk Score" of the nodes traversed should decrease.

```python
def calculate_metrics(graph, route):
    if not route: return 0, 0
    #calculate distance
    dist = nx.path_weight(graph, route, weight='length')
    #calculate risk
    risk = sum(graph.nodes[n].get('risk_score', 0) for n in route)
    return int(dist), risk

short_riskNode_sum = 0
safe_riskNode_sum = 0

# metrics for both routes
l_short, r_short = calculate_metrics(G, route_shortest)
l_safe, r_safe   = calculate_metrics(G, route_safest)

print("------------ Test Case 2 ------------")
print(f"\nAlpha: {ALGORITHM_ALPHA}")
print(f"SHORTEST PATH:           Distance = {l_short}m")
for node in route_shortest:
    risk = G.nodes[node].get('risk_score', 0)
    if risk > 0:
        print(f"->Node {node} has a risk of {risk}!")
        short_riskNode_sum += 1
print(f"This path has {short_riskNode_sum} risk nodes!")
print(f"\n--- Total Risk Score               =  {r_short} ---")
print("-" * 45)

print(f"SAFEST PATH:              Distance = {l_safe}m")
for node in route_safest:
    risk = G.nodes[node].get('risk_score', 0)
    if risk > 0:
        print(f"->Node {node} has a risk of {risk}!")
        safe_riskNode_sum += 1
print(f"This path has {safe_riskNode_sum} risk nodes!")
print(f"\n--- Total Risk Score                =  {r_safe} ---")

print("-" * 45)
print(f"Avoided {r_short - r_safe} points of risk!!")
print(f"Cost: Needs a {l_safe - l_short}m detour.")
```

**Actual Result**:

Alpha 0: Shorter. Higher Alpha: Longer paths, lower risks. The agent successfully diverted around risk nodes, that can be identified in the heatmap, confirming the cost function successfully penalises high risk edges.
It's also important to note, however, that local crime density and selected Start and Goal (Destination) points affect the "usefullness" or range of available paths with more or less crime data/safety calculations. For example, at a certain point, Alpha will no longer affect the path if it finds a "perfectly safe" path.

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Shortest Path', 'Safest Path']
risk_scores = [r_short, r_safe]
distances = [l_short, l_safe]

x = np.arange(len(categories))
width = 0.35

fig, ax1 = plt.subplots(figsize=(8, 6))

#plot
color_risk = 'tab:blue'
ax1.set_xlabel('Route Type', fontweight='bold')
ax1.set_ylabel('Total Risk Score', color=color_risk, fontweight='bold')
bars1 = ax1.bar(x - width/2, risk_scores, width, color=color_risk, alpha=0.8, label='Risk Score')
ax1.tick_params(axis='y', labelcolor=color_risk)
ax1.set_ylim(0, max(risk_scores) * 1.3)

ax2 = ax1.twinx()
color_dist = 'tab:red'
ax2.set_ylabel('Distance (meters)', color=color_dist, fontweight='bold')
bars2 = ax2.bar(x + width/2, distances, width, color=color_dist, alpha=0.8, label='Distance')
ax2.tick_params(axis='y', labelcolor=color_dist)
ax2.set_ylim(0, max(distances) * 1.3)

def autolabel(bars, ax, color_text):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{int(height)}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom',
                    color=color_text, fontweight='bold')

autolabel(bars1, ax1, color_risk)
autolabel(bars2, ax2, color_dist)

plt.title('Evaluation: Risk Reduction vs. Distance Cost', fontsize=14, pad=20)
ax1.set_xticks(x)
ax1.set_xticklabels(categories)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)

plt.grid(axis='y', linestyle='--', alpha=0.3)
fig.tight_layout()
plt.show()
```

#### Effectiveness in Solving the Problem

The primary goal of "SafeWalk" was to calculate a walking path that minimizes exposure to crime risk while maintaining a reasonable travel distance, by incorporating real world crime data into pathfinding. My implementation successfully achieved this objective, generating, the scale depending on `ALGORITHM_ALPHA`, two contrasting routes: the shortest distance path and a safer path that penalises and thus avoids areas with higher crime.

The results show that the model actually differentiates between these objectives. The shortest route prioritises minimal distance, often passing through areas with dense crime clusters, whereas the safest route deliberately avoids these regions, even at the cost of increased travel distance. This confirms that the algorithm is not just producing redundant differences, or creating new paths at random, but is effectively and deliberately embedding domain relevant safety information into pathdfinding.



#### Performance Evaluation

From a computational perspective, the system performs efficiently. The A* search consistently completes in under a few of seconds for graphs of more than 18,000 nodes. This performance would not be achievable using an uninformed search strategy, which would explore significantly more nodes.

Although the routing graph is localised, the underlying crime dataset contains over one million records covering the full Metropolitan Police (MET) area. Preprocessing and aggregation ensure that this scale does not negatively impact runtime performance. For example, to prevent bottlenecks when processing the large crime dataset, I avoided iterative loops and looked for another solution, which turned out to be vectorised operations. Specifically, I utilized osmnx's `nearest_nodes`, which employs a "cKDTree" to map all crime coordinates to graph nodes in a single batch query (Boeing, G., 2025). Furthermore, risk scores were aggregated using Pandas' vectorized `groupby` and `sum` functions, ensuring the pre processing  remained efficient and scalable.

This means that the system can actually handle large real world datasets while remaining responsive and useful. This makes the solution suitable for interactive or near real time use: it could be further developed into something truly useful.



#### Evaluation of Results in the Domain Context

In the context of urban mobility and pedestrian safety, my model results highlight a necessary trade off: safer routes are often longer routes. The safest path frequently avoids busy streets with high reported crime and instead selects different and indirect alternatives.

This shows an important insight, that statistical safety does not always align with perceived safety. For example, while the algorithm may classify a quiet residential street as safer due to lower reported crime, a real pedestrian might feel less safe due to reduced visibility or unfamiliarity.

#### Insights and Potential Improvements

The solution I worked on provides several insights:

* Crime is often concentrated at specific nodes such as transport options (buses, trains), bars, arcades, popular hotspots. This suggests that targeted  interventions (e.g., improved lighting at night/dark, more visible CCTV, or even patrols) could be more effective at increasing safety and lowering crime rate here.

* A route suitable during daytime may be inappropriate at night. The user controlled weighting (Alpha) parameter already partially addresses this by allowing the user to trade efficiency for safety depending on context.



#### Strengths and Limitations of the Implementation
Strengths

* The A* search algorithm is fully transparent. Pathfinding decisions can be directly explained through the cost function, allowing clear justification for detours based on crime severity. This is a big advantage over black-box AI approaches, for example.

* The system dynamically retrieves map data using `OpenStreetMap` and can be applied to any area covered by the MET police, and with minimal modification, to other urban regions. Despite operating on a crime dataset exceeding one million entries, the solution maintains low processing time.

* The adjustable safety weighting allows the system to accommodate different user needs, increasing its practical relevance and flexibility, not just for testing.

Limitations

* The model relies on reported crime data, which may overrepresent areas with higher police presence and underrepresent unreported incidents. This can skew pathfinding decisions away from well policed but potentially safer areas, and vice versa- safety depends on quality of gathered crime data, which is out of my hands and depends on the police.

* Crime is treated as temporally static. The model does not distinguish between daytime and night-time risk, which can lead to irrational avoidance of locations that are safe under certain conditions, as aforementioned.

* Mapping crimes to the nearest node may over-penalise some road segments, occasionally resulting in inefficient detours. Some alternative paths may not exist if the Start and Goal are too close, or there aren't viable options in the Alpha Zero test either, etc.

* As mentioned previously during Testing, at some point the Alpha value may no longer affect the created path based on the quality of the crime data. Further highlighting the imporance of accurate crime data.

## Part E: Equity, Diversity, Inclusion (EDI) and Sustainability (10 Marks)

### 1. Equity, Diversity, and Inclusion (EDI) (5 Marks)

**Bias & Fairness**

The most significant ethical risk associated with this implementation is algorithmic bias originating from the input data. The system relies on Police.uk crime data, which reflects reported and recorded crime, rather than true crime prevalence.

Crime datasets are not neutral, and not 100% accurate. It's a sad reality that many crimes are simply not reported, and historical and structural factors mean that marginalised and minority communities are often subject to disproportionately high levels of policing and surveillance, including stop-and-search practices. As a result, these areas may appear as “high risk” clusters in the dataset due to higher detection and reporting rates, while crime in more affluent areas may be under reported, for example. The AI therefore risks learning patterns of policing intensity, not objective safety.

If deployed at scale, by consistently routing pedestrians away from certain neighbourhoods, the application could reduce footfall, negatively impact local businesses, and increase social isolation. An area is labelled as unsafe by the algorithm, people avoid it, natural surveillance decreases, and the area may become genuinely less safe. In this way, an AI system designed to promote safety could unintentionally reinforce existing social and economic inequalities.

**Accessibility and Inclusion**

Fear of crime disproportionately affects women, the elderly, children, and other vulnerable groups, often limiting our mobility, particularly after dark. By quantifying safety and offering transparent route alternatives, the system lets users make more informed decisions and have access to urban spaces they might otherwise avoid. This is a tangible inclusion benefit from which people would benefit from and enjoy.

However, applying this model and solution assumes access to a smartphone or at  the very least, the internet. People from much lower income backgrounds, who may also live in higher crime areas, are less likely to have the resources required to benefit from this. Without careful deployment, the model would risk not serving users who would benefit from it the most, in a real life situation.

### 2. Environmental Sustainability (5 Marks)

From a sustainability perspective, I believe my prototype actually aligns strongly with the principles of Green AI (Schwartz, R. et al., 2019), which emphasises energy efficient model design over computationally intensive approaches.

The decision to use A* heuristic search rather than a deep learning model (such as an RNN) significantly reduces environmental impact. Training neural networks on millions of data points typically requires very high performance GPUs operating for extended periods, resulting in a substantially higher carbon footprint. In contrast, A* is deterministic, lightweight, and requires no training phase, making it far more energy efficient.

Further, as demonstrated in Part C, the use of vectorised operations reduced preprocessing time for the 1 million+ crime records from hours to seconds. This optimisation, apart from being a better experience whilst coding and testing, and for the user, it directly lowers energy consumption by minimising CPU runtime and avoiding long high load processing.

Compared to other scopes of interest or applications, the system relies on compact, structured data formats (CSV crime data and OpenStreetMap graphs), which require only megabytes of storage. This is significantly more sustainable than other systems that are based on, for example, continuous video feeds or computer vision, which can require terabytes of storage and constant processing.

For real world deployment, I believe that the environmental impact could be  reduced by implementing caching and incremental updates, on for example an app. Frequently requested routes and pre computed heuristic distances could be cached, avoiding repeated pathfinding computations for the same queries. This would significantly reduce unnecessary recomputation, CPU usage, and energy consumption at scale, particularly in dense urban areas where the app could be used.
